```python
#import module and packages to use

from tkinter import *
from PIL import Image, ImageTk
import webbrowser
import datetime
import os

#defining functionalities

def get_response(question):
    default_response = "I'm sorry, I don't understand that question."
    
    
               
    if "youtube" in question.lower():
        webbrowser.open("https://www.youtube.com/")
        return "Opening YouTube in your web browser."
    elif "google" in question.lower():
        query = question.lower().replace("wikipedia", "").strip()
        webbrowser.open(query)
        return "Opening query in your web browser."
    
    elif "date" in question.lower():
        # Get the current date
        current_date = datetime.datetime.now().strftime("%Y-%m-%d")
        response = f"The current date is {current_date}."
        return response

    elif "time" in question.lower():
        # Get the current time
        current_time = datetime.datetime.now().strftime("%H:%M:%S")
        response = f"The current time is {current_time}."
        return response

    elif "calculate" in question.lower() or "solve" in question.lower():
        
            # Extract the mathematical expression from the question
            expression = question.lower().replace("calculate", "").replace("solve", "").strip()
            # Evaluate the expression using eval() and round the result to 2 decimal places
            result = eval(expression)
            response = f"The result is: {round(result, 2)}"
            return response
    elif "search files" in question.lower():
        search_term = question.lower().replace("search files", "").strip()
        matched_files = search_files(search_term)
        response = "Matching files found:\n" + "\n".join(matched_files) if matched_files else "No files found."
        return response
           


    else:    
        with open("responses.txt", "r") as file:
            for line in file:
                q, response = line.strip().split("|")
                if q.lower() in question.lower():
                    return response
    return default_response


def botReply():
    question = que_Entry.get()
    textarea.tag_config("bot", foreground="skyblue")
    textarea.insert(END, "YOU: " + question + "\n")
    response = get_response(question)
    textarea.insert(END, "BOT: " + response + "\n","bot")
    que_Entry.delete(0, END)
    textarea.see(END)



def search_files(search_term):
    matched_files = []
    for root, dirs, files in os.walk("."):  # Replace "." with the root directory path you want to search
        for file in files:
            if search_term in file:
                matched_files.append(os.path.join(root, file))
    return matched_files
root = Tk()



root.geometry("500x570+100+30")
root.title("Chatbot created by jugraj")
root.config(bg ="skyblue")
# Load the JPG image using Pillow
img = ImageTk.PhotoImage(Image.open("chatbotlogo2.png"))
labl1 = Label(root,image = img,bg = "skyblue")
labl1.pack(pady= 10)


centreframe = Frame(root)
centreframe.pack()

scrollbar = Scrollbar(centreframe)
scrollbar.pack(side=RIGHT)


scrollbar = Scrollbar(centreframe)
scrollbar.pack(side=RIGHT)

textarea = Text(centreframe,font = ("Georgia",20,"bold"),height = 10, yscrollcommand=scrollbar.set,wrap = "word")
textarea.pack(side = LEFT)

scrollbar.config(command=textarea.yview)


que_Entry = Entry(root,font=("verdana",20,"bold"))
que_Entry.pack(pady= 5,fill = X)

imgbutton = ImageTk.PhotoImage(Image.open("askicon.png"))

askButton = Button(root,image = imgbutton,command=botReply)
askButton.pack()

root.mainloop()
```
